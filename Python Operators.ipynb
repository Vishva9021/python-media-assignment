```python
# 1. Calc$late the sum, difference, product, and quotient of two numbers.
a = int(input("Enter the first Number :"))
b = int(input("Enter the second Number : "))
x = a+b
z = a-b
y = a*b
s = a/b
print("Addition:",x)
print("subtraction :",y)
print("Multiplication :",z)
print("Division :",s)

# 2. Perform various assignment operations on a variable.
# Initialize a variable
x = 10
print("Initial value of x:", x)

# Assignment: Assign a new value to the variable
x = 20
print("After assignment, x is now:", x)

# Addition assignment: Add a value to the variable and assign the result back to the variable
x += 5
print("After addition assignment, x is now:", x)

# Subtraction assignment: Subtract a value from the variable and assign the result back to the variable
x -= 3
print("After subtraction assignment, x is now:", x)

# Multiplication assignment: Multiply the variable by a value and assign the result back to the variable
x *= 2
print("After multiplication assignment, x is now:", x)

# Division assignment: Divide the variable by a value and assign the result back to the variable
x /= 4
print("After division assignment, x is now:", x)

# Floor division assignment: Perform floor division on the variable and assign the result back to the variable
x //= 3
print("After floor division assignment, x is now:", x)

# Exponentiation assignment: Raise the variable to a power and assign the result back to the variable
x **= 2
print("After exponentiation assignment, x is now:", x)

# Modulus assignment: Perform modulus operation on the variable and assign the result back to the variable
x %= 5
print("After modulus assignment, x is now:", x)


# 3. Compare two numbers and print the results.
# Define two numbers
number1 = 10
number2 = 20

# Compare the two numbers
if number1 > number2:
    print(f"{number1} is greater than {number2}")
elif number1 < number2:
    print(f"{number1} is less than {number2}")
else:
    print(f"{number1} is equal to {number2}")

# 4. Check conditions using logical operators.
# Define two numbers
number1 = 10
number2 = 20
number3 = 5

# Check conditions using logical operators
if number1 < number2 and number1 < number3:
    print(f"{number1} is less than both {number2} and {number3}")
elif number1 < number2 or number1 < number3:
    print(f"{number1} is less than either {number2} or {number3}")
elif not (number1 > number2):
    print(f"{number1} is not greater than {number2}")

# 5 .Check the identity of variables

# Define variables
x = [1, 2, 3]
y = [1, 2, 3]
z = x

# Check identity of variables
print("x is y:", x is y)  # False, because x and y are distinct objects even though they have the same values
print("x is z:", x is z)  # True, because z refers to the same object as x

# 6. Perform bitwise operations on any two integers.
# Define two integers
a = 10   # Binary: 1010
b = 5    # Binary: 0101

# Bitwise AND
result_and = a & b
print("Bitwise AND of", a, "and", b, "is", result_and)  # Output: 0 (Binary: 0000)

# Bitwise OR
result_or = a | b
print("Bitwise OR of", a, "and", b, "is", result_or)    # Output: 15 (Binary: 1111)

# Bitwise XOR
result_xor = a ^ b
print("Bitwise XOR of", a, "and", b, "is", result_xor)  # Output: 15 (Binary: 1111)

# Bitwise NOT (complement) of a
result_not_a = ~a
print("Bitwise NOT of", a, "is", result_not_a)          # Output: -11 (Binary: 11110101 in two's complement form)

# Bitwise left shift a by 2 bits
result_left_shift = a << 2
print("Bitwise left shift of", a, "by 2 bits is", result_left_shift)  # Output: 40 (Binary: 101000)

# Bitwise right shift a by 2 bits
result_right_shift = a >> 2
print("Bitwise right shift of", a, "by 2 bits is", result_right_shift)  # Output: 2 (Binary: 10)


# 7. Use unary operators to change the sign of a number.
# Define a number
number = 10

# Change the sign of the number using unary operator
number = -number

print("Number after changing the sign:", number)  # Output: -10

# 8. Use the ternary operator to assign values based on conditions.
# Define a condition
condition = True

# Assign values based on the condition using the ternary operator
x = 10 if condition else 20

# Print the value of x
print("Value of x based on the condition:", x)
```
